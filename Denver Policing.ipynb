```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv("denver-1.nov.csv")
```

```python
data.head()
```

```python
data["disposition"].unique()
```

```python
type(data.loc[0,"time"])
```

```python
def late_at_night(time_stamp):
    hour = int(time_stamp[0:2])
    if hour < 4 or hour > 21:
        return True
    else:
        return False
```

```python
late_at_night(data.loc[0,"time"])
```

```python
data["Late_At_Night"] = data["time"].apply(late_at_night)
```

```python
data.head()
```

```python
type(data.loc[0,"arrest_made"])
```

```python
def fix_booleans(boolean):
    if boolean == True:
        return 1
    else:
        return 0
```

```python
data["arrest_made"] = data["arrest_made"].apply(fix_booleans)
```

```python
data["citation_issued"] = data["citation_issued"].apply(fix_booleans)
```

```python
data["warning_issued"] = data["warning_issued"].apply(fix_booleans)
```

```python
data.head()
```

```python
gb = data.groupby("Late_At_Night").agg({"arrest_made":np.mean, "citation_issued":np.mean, "warning_issued":np.mean})
```

```python
gb
```

```python
gb.reset_index(inplace=True)
```

```python
gb["arrest_made"] = gb["arrest_made"].apply(lambda x: round(x*100,2))
```

```python
gb["citation_issued"] = gb["citation_issued"].apply(lambda x: round(x*100,2))
```

```python
gb["warning_issued"] = gb["warning_issued"].apply(lambda x: round(x*100,2))
```

```python
c1 = alt.Chart(gb, title = "Percent of Stops with an Arrest Based on Time of Day").mark_bar().encode(x="Late_At_Night", 
                                     y="arrest_made", 
                                     color="Late_At_Night", 
                                     tooltip = ["arrest_made"])
c2 = alt.Chart(gb, title = "Percent of Stops with a Citation Based on Time of Day").mark_bar().encode(x="Late_At_Night", 
                                     y="citation_issued", 
                                     color="Late_At_Night", 
                                     tooltip = ["citation_issued"])
c3 = alt.Chart(gb, title = "Percent of Stops with a Warning Based on Time of Day").mark_bar().encode(x="Late_At_Night", 
                                     y="warning_issued", 
                                     color="Late_At_Night", 
                                     tooltip = ["warning_issued"])
c1 | c2 | c3
```
